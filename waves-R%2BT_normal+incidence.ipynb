### Reflections and transmissions of EM waves at a boundary
This brief simulation helps us develop an intuitive picture for how the amplitude of a wave changes as it is reflected or transmitted from a boundary.
Griffiths section 9.3.2

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

```python
# define a range of indices of refraction
n2 = np.arange(1,4,0.1)
n1 = 1;

beta = n2/n1

E_0R = (1-beta)/(1+beta)

E_0T = 2/(1+beta)
```

```python
plt.plot(n2,np.abs(E_0R),'r',label = 'reflection')    
plt.plot(n2,np.abs(E_0T),'b',label = 'transmission')    

plt.xlabel('index of refraction n2',fontsize=15)
plt.ylabel('amplitude $E/E_{0I}$',fontsize=15)
plt.title('refl. and trans. E field amplitudes from a boundary\nn1 = 1, n2 varies',fontsize=15)
plt.grid(True)
# plt.ylim([1e15,np.max(n_Phos)*10])

plt.legend(frameon=False,loc='best') 
plt.tight_layout() 
```

```python
R = np.abs(E_0R)**2
T = beta*np.abs(E_0T)**2
# T = 4*n1*n2/(n1+n2)**2

plt.plot(n2,R,'r',label = 'reflection')    
plt.plot(n2,T,'b',label = 'transmission')    
plt.plot(n2,R+T,'g',label = 'R+T')    

plt.xlabel('index of refraction n2',fontsize=15)
plt.ylabel('intensity $I/I_I$',fontsize=15)
plt.title('refl. and trans. E field amplitudes from a boundary\nn1 = 1, n2 varies',fontsize=15)
plt.grid(True)
# plt.ylim([1e15,np.max(n_Phos)*10])

plt.legend(frameon=False,loc='best') 
plt.tight_layout() 
```

```python
Rval = np.interp(1.5,n2,R)
print(Rval)
```

```python
from IPython.html.widgets import interact

def plot_sine(amplitude=1.0, k = 1.0, omega=1.0, t = 1.0):
    plt.ylim(-1.0, 1.0);
    x = np.linspace(0, 10, 1000)
    plt.plot(x, amplitude*np.sin(x*k-omega*t));
    plt.xlabel('location (m)',fontsize=15)
    plt.ylabel('electric field (a.u.)',fontsize=15)

interact(plot_sine, amplitude=(0.0, 1.0),k=(0.0, 5.0),omega=(0.5, 10.0),t=(0.0, 5.0));
```
